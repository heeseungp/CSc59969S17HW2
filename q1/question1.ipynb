# Part 1: Modify the plot template to plot two of the above measuremeants against each other. Do not forget to update labels!

```python
%matplotlib inline
import numpy as np
import matplotlib


import matplotlib.pyplot as plt
import matplotlib as mpl


data = np.recfromcsv('crabs.csv', names = "f0, sp, sex, index, fl, rw, cl, cw, bd", usecols=('rw', 'bd', 'fl'))
rw = [] #rear width (mm)
bd = [] #body depth (mm)
fl = [] #frontal lobe size (mm)

for item in data[1:]:
    rw.append(float(item[0]))
    bd.append(float(item[1]))
    fl.append(float(item[2]))
    
#first create the figure
fig = plt.figure(figsize=(8,5))

#create a grid of 1 row and 1 column for the plot
#[left, bottom, width, height]
gs = mpl.gridspec.GridSpec(1,1)
#put a plot in the first row, first column
ax = fig.add_subplot(gs[0])

#Have only two spines visible and set properties
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

#set tick properties
ax.xaxis.set_tick_params('major', length=5, labelsize=7)
ax.yaxis.set_tick_params('major', length=5, labelsize=7)

#plot data and customize look of plot
ax.plot(rw,bd, marker='.', color='k', linestyle='',
        markersize=4, clip_on=False)

#set axis limits 
ax.set_xlim((5,22))
ax.set_ylim((5,22))

#scale the view to show all datapoints
ax.autoscale_view()

#set labels
ax.set_xlabel(r'Rear Width (mm)', fontsize=10, family='sans-serif')
ax.set_ylabel('Body Depth (mm)', fontsize=10, family='sans-serif')

fig.savefig('question1a.svg')
```

## Part 2b: Represent the third variable with the circles of different sizes (Hint: use markersize keyword argument of plot)

```python
%matplotlib inline
import numpy as np
import matplotlib


import matplotlib.pyplot as plt
import matplotlib as mpl


data = np.recfromcsv('crabs.csv', names = "f0, sp, sex, index, fl, rw, cl, cw, bd", usecols=('rw', 'bd', 'fl', 'sp'))
rw = [] #rear width (mm)
bd = [] #body depth (mm)
fl = [] #frontal lobe size (mm)
sp = [] #color (blue or orange)
for item in data[1:]:
    rw.append(float(item[0]))
    bd.append(float(item[1]))
    fl.append(float(item[2]))
    sp.append((item[3]).decode("utf-8") )
    
#first create the figure
fig = plt.figure(figsize=(8,5))

#create a grid of 1 row and 1 column for the plot
#[left, bottom, width, height]
gs = mpl.gridspec.GridSpec(1,1)
#put a plot in the first row, first column
ax = fig.add_subplot(gs[0])

#Have only two spines visible and set properties
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

#set tick properties
ax.xaxis.set_tick_params('major', length=5, labelsize=7)
ax.yaxis.set_tick_params('major', length=5, labelsize=7)

#plot data and customize look of plot
for idx, val in enumerate(fl):
    ax.plot(rw[idx],bd[idx], marker='.', color='k', linestyle='',
            markersize=val-5, clip_on=False)

#set axis limits 
ax.set_xlim((5,22))
ax.set_ylim((5,22))

#scale the view to show all datapoints
ax.autoscale_view()

#set labels
ax.set_xlabel(r'Rear Width (mm)', fontsize=10, family='sans-serif')
ax.set_ylabel('Body Depth (mm)', fontsize=10, family='sans-serif')

fig.savefig('question2b.svg')
```

## Part 3: Group variables belonging to the same crab group (blue male, blue female, etc., choose one method):
by connecting the points with a line


```python
%matplotlib inline
import numpy as np
import matplotlib


import matplotlib.pyplot as plt
import matplotlib as mpl


data = np.recfromcsv('crabs.csv', names = "f0, sp, sex, index, fl, rw, cl, cw, bd", usecols=('rw', 'bd', 'fl', 'sp'))
rw = [] #rear width (mm)
bd = [] #body depth (mm)
fl = [] #frontal lobe size (mm)
sp = [] #color (blue or orange)
for item in data[1:]:
    rw.append(float(item[0]))
    bd.append(float(item[1]))
    fl.append(float(item[2]))
    sp.append((item[3]).decode("utf-8") )
    
#first create the figure
fig = plt.figure(figsize=(8,5))

#create a grid of 1 row and 1 column for the plot
#[left, bottom, width, height]
gs = mpl.gridspec.GridSpec(1,1)
#put a plot in the first row, first column
ax = fig.add_subplot(gs[0])

#Have only two spines visible and set properties
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

#set tick properties
ax.xaxis.set_tick_params('major', length=5, labelsize=7)
ax.yaxis.set_tick_params('major', length=5, labelsize=7)

blue_rw = []
blue_bd = []

orange_rw = []
orange_bd = []

for i in range(0,100):
    blue_rw.append(rw[i])
    blue_bd.append(bd[i])
for j in range(101,200):
    orange_rw.append(rw[j])
    orange_bd.append(bd[j])
    
#plot data and customize look of plot

ax.plot(blue_rw,blue_bd, marker='.', color='b', linestyle='-',
            markersize=6, clip_on=False)
ax.plot(orange_rw,orange_bd, marker='.', color='m', linestyle='-',
            markersize=6, clip_on=False)
    
#set axis limits 
ax.set_xlim((5,22))
ax.set_ylim((5,22))

#scale the view to show all datapoints
ax.autoscale_view()

#set labels
ax.set_xlabel(r'Rear Width (mm)', fontsize=10, family='sans-serif')
ax.set_ylabel('Body Depth (mm)', fontsize=10, family='sans-serif')

fig.savefig('question3.svg')
```
