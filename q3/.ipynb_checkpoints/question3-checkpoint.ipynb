# Part A

1.Convert the temperatures to Fahrenheit from Kelvin so they are more easily interpretable to an American audience.

2.Make a box and whiskers diagram where the Y axis is labeled by temperature in degrees Fahrenheit and the X axis is labeled by city. Since you don't want the city names to overlap, rotate the text so they run vertically. Use all the temperatures. Write a caption pointing out what interesting things to interpret from the cities.

3.In addition to the previous diagram, make a separate diagrams for the first 20,000 measurements and the last 20,000 measurements. Put them in a single figure (two separate subplots). Use the add_sublot for this. Label/title the total graph and subplots clearly. Do you see any trend?

```python
from scipy.io import loadmat
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt


#formula to convert from kelvin to fahrenheit
#http://thelivingpearl.com/2014/01/02/temperature-conversation-application-in-python/
def kToF(list_temp):
    newList = []
    for idx, element in enumerate(list_temp):
        newList.append((element[0]*9/5.0)-459.67)
    return newList


johannesburg = loadmat('Johannesburg.mat')
paris = loadmat('Paris.mat')
saopaulo = loadmat('SaoPaulo.mat')
singapore = loadmat('Singapore.mat')
sydney = loadmat('Sydney.mat')
vancouver = loadmat('Vancouver.mat')

jo_temp = kToF(johannesburg['temperature'])
pa_temp = kToF(paris['temperature'])
sa_temp = kToF(saopaulo['temperature'])
si_temp = kToF(singapore['temperature'])
sy_temp = kToF(sydney['temperature'])
va_temp = kToF(vancouver['temperature'])

data_to_plot = [jo_temp, pa_temp, sa_temp, si_temp, sy_temp, va_temp]

fig = plt.figure(1,figsize=(10,8))
ax = fig.add_subplot(111)
bp = ax.boxplot(data_to_plot)
ax.set_xticklabels(['Johannesburg', 'Paris', 'Sao Paulo', 'Singapore', 'Sydney', 'Vancouver'])
ax.set_xlabel('City')
ax.set_ylabel('Temperature (F)')
ax.set_title('Meaurements of temperature from 1980 to 2009 in Different Cities')
plt.show()
```

# Part B: Pandas and Trend Design Problem
We would like to see if there has been any change over time in the temperatures of the city. Fluctuations may make it hard to see differences. For example a small but steady increase in average winter or summer temperatures, may be masked by hourly fluctuations. Working with dates can be rather tricky. Convert the mat data into Pandas data structures:

http://pandas.pydata.org/

Check out the tutorials on Pandas.

1. Define Winter as December, January and Februrary; Spring as March, April and May; Summer as June, July and August and Fall as September, October and November. Separately for each year (1948-2008) and each city compute the average temperature for that seasons.
2. Design and create a graph(s) that shows the seasonal trends (or lack of trends) over the years from 1948 to 2008 (last year is incomplete).
3. Use small multiples with subplot and/or GridSpec to put multiple plots together for comparison
4. Use linear regresion lines and corelation coeefficients to show existance or lack of trends over time

It is up to you how you will group the graphs. You may do it by city or by season but try to show trends or lack of trends in as compact a way as possible (high data to ink ratio) while supporting many visual queries. Visual queiries should include
- Are there different trends for different seasons?
- How do city compare? Is there more or less of a change for tropical cities then temperate cities?

Be careful not to washout changes by, for example, grouping different cities together when computing changes. Also consider using filled area under the curve rather than a straight line plot (if appropriate). Also consider you scale carefully. Remember that the 0 of temperature is somewhat arbitrary. Often climatolgists will subtract an average over a series to show differences more clearly. When displaying the trend with the total average subtracted we call this the "temperature anomaly" and label as such. In the caption or somewhere else should clearly describe how the data has been processed (which mean is subtracted, what averages are shown).

Note: Your graphs should be understandable to anyone reading them. Look at them carefully assuming the person viewing them did not read this homework assignment.

```python
from scipy.io import loadmat
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt


#formula to convert from kelvin to fahrenheit
#http://thelivingpearl.com/2014/01/02/temperature-conversation-application-in-python/
def kToF(list_temp):
    newList = []
    for idx, element in enumerate(list_temp):
        newList.append((element[0]*9/5.0)-459.67)
    return newList
    
def computeEachYr(time, temperature, season):
    
    eachyr = []
    eachyr_element = 0
    
    eachtemp = []
    temporary_temp_list = []
    
    for idx, val in enumerate(time):
        if int(val[0:4]) >= 1948 and int(val[0:4]) <= 2008:
            if eachyr_element != int(val[0:4]):
                eachyr_element = int(val[0:4])
                eachyr.append(eachyr_element)
                if len(temporary_temp_list) != 0:
                    eachtemp.append(temporary_temp_list)
                    temporary_temp_list = []
            if val[5:7] in season:
                temporary_temp_list.append(temperature[idx])
                
    for idx, val in enumerate(eachtemp):
        eachtemp[idx] = (sum(val)/len(val))
        
    return eachyr, eachtemp
    
winter = ['12','01','02']
spring = ['03','04','05']
summer = ['06', '07','08']
fall = ['09','10','11']
            
city_name = ['johannesburg', 'paris', 'saopaulo', 'singapore', 'sydney', 'vancouver']

johannesburg = loadmat('Johannesburg.mat')
paris = loadmat('Paris.mat')
saopaulo = loadmat('SaoPaulo.mat')
singapore = loadmat('Singapore.mat')
sydney = loadmat('Sydney.mat')
vancouver = loadmat('Vancouver.mat')

jo_temp = kToF(johannesburg['temperature'])
pa_temp = kToF(paris['temperature'])
sa_temp = kToF(saopaulo['temperature'])
si_temp = kToF(singapore['temperature'])
sy_temp = kToF(sydney['temperature'])
va_temp = kToF(vancouver['temperature'])

jo_time = johannesburg['time']
pa_time = paris['time']
sa_time = saopaulo['time']
si_time = singapore['time']
sy_time = sydney['time']
va_time = vancouver['time']

jo_yr_winter, jo_temp_winter = computeEachYr(jo_time, jo_temp, winter)
pa_yr_winter, pa_temp_winter = computeEachYr(pa_time, pa_temp, winter)
sa_yr_winter, sa_temp_winter = computeEachYr(sa_time, sa_temp, winter)
si_yr_winter, si_temp_winter = computeEachYr(si_time, si_temp, winter)
sy_yr_winter, sy_temp_winter = computeEachYr(sy_time, sy_temp, winter)
va_yr_winter, va_temp_winter = computeEachYr(va_time, va_temp, winter)
                  
jo_yr_spring, jo_temp_spring = computeEachYr(jo_time, jo_temp, spring)
pa_yr_spring, pa_temp_spring = computeEachYr(pa_time, pa_temp, spring)
sa_yr_spring, sa_temp_spring = computeEachYr(sa_time, sa_temp, spring)
si_yr_spring, si_temp_spring = computeEachYr(si_time, si_temp, spring)
sy_yr_spring, sy_temp_spring = computeEachYr(sy_time, sy_temp, spring)
va_yr_spring, va_temp_spring = computeEachYr(va_time, va_temp, spring)
                  
jo_yr_summer, jo_temp_summer = computeEachYr(jo_time, jo_temp, summer)
pa_yr_summer, pa_temp_summer = computeEachYr(pa_time, pa_temp, summer)
sa_yr_summer, sa_temp_summer = computeEachYr(sa_time, sa_temp, summer)
si_yr_summer, si_temp_summer = computeEachYr(si_time, si_temp, summer)
sy_yr_summer, sy_temp_summer = computeEachYr(sy_time, sy_temp, summer)
va_yr_summer, va_temp_summer = computeEachYr(va_time, va_temp, summer)
                  
jo_yr_fall, jo_temp_fall = computeEachYr(jo_time, jo_temp, fall)
pa_yr_fall, pa_temp_fall = computeEachYr(pa_time, pa_temp, fall)
sa_yr_fall, sa_temp_fall = computeEachYr(sa_time, sa_temp, fall)
si_yr_fall, si_temp_fall = computeEachYr(si_time, si_temp, fall)
sy_yr_fall, sy_temp_fall = computeEachYr(sy_time, sy_temp, fall)
va_yr_fall, va_temp_fall = computeEachYr(va_time, va_temp, fall)

winter_yr_collection = [jo_yr_winter, pa_yr_winter, sa_yr_winter, si_yr_winter, sy_yr_winter, va_yr_winter]
spring_yr_collection = [jo_yr_spring, pa_yr_spring, sa_yr_spring, si_yr_spring, sy_yr_spring, va_yr_spring]
summer_yr_collection = [jo_yr_summer, pa_yr_summer, sa_yr_summer, si_yr_summer, sy_yr_summer, va_yr_summer]
fall_yr_collection = [jo_yr_fall, pa_yr_fall, sa_yr_fall, si_yr_fall, sy_yr_fall, va_yr_fall]

winter_temp_collection = [jo_temp_winter, pa_temp_winter, sa_temp_winter, si_temp_winter, sy_temp_winter, va_temp_winter]
spring_temp_collection = [jo_temp_spring, pa_temp_spring, sa_temp_spring, si_temp_spring, sy_temp_spring, va_temp_spring]
summer_temp_collection = [jo_temp_summer, pa_temp_summer, sa_temp_summer, si_temp_summer, sy_temp_summer, va_temp_summer]
fall_temp_collection = [jo_temp_fall, pa_temp_fall, sa_temp_fall, si_temp_fall, sy_temp_fall, va_temp_fall]

fig = plt.figure(figsize=(10,7))
gs = mpl.gridspec.GridSpec(1,1)
ax = fig.add_subplot(gs[0])
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_tick_params('major', length=2, labelsize=6)
ax.yaxis.set_tick_params('major', length=2, labelsize=6)
ax.set_color_cycle(['red', 'black', 'yellow', 'blue', 'magenta', 'green'])

for idx, val in enumerate(winter_yr_collection):
    ax.plot(val[:-1], winter_temp_collection[idx], marker='.', linestyle='-',
        markersize=4, clip_on=False, label = city_name[idx])

#set axis limits 
ax.set_xlim((min(jo_yr_winter),max(jo_yr_winter)))
ax.set_ylim((10,90))
#scale the view to show all datapoints
ax.autoscale_view()
#set labels
ax.set_xlabel('Year', fontsize=10, family='sans-serif')
ax.set_ylabel('Temperature (F)', fontsize=10, family='sans-serif')
# plt.xticks(x, labels, rotation='vertical')
ax.set_title('Average temperature of Winter Season in Each Different Cities')
plt.legend(ncol = 1, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```
