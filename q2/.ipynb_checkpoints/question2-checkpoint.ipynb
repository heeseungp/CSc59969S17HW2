```python
import numpy as np
from scipy.io import loadmat
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy.stats

data = loadmat('AQUA.2013089.2055.mat')
```

```python
n = data['B1'].shape[0] * data['B1'].shape[1]
spercentage = 0.45
numSamps = int(n * spercentage)

samplesInds = list(range(1,n))
np.random.shuffle(samplesInds)
samplesInds = samplesInds[0:numSamps]
```

```python
b1 = data['B1']
x_b1 = b1.ravel()[samplesInds]
b6 = data['B6']
y_b6 = b6.ravel()[samplesInds]
```

One thing we like to do is compare with a linear regression. If the data fits a line then there should be a good corrolation between line fit and the data.

Use the polyfit function to fit the B1 samples vs. the B6 samples to get a line fit.

Use polyval (on two x values ... one just below the min (90%) and one just above max (110%)) to get a bet fit line.

Show the line (plot) in red on top of the samples.

Make sure you have a legend that says "Linear regression line"

Compute the slope and use matplotlibs "text" function to put a label "slope = " (the slope value) on the graph. You will need to manually adjust the position so it doesn't block other elements.

From scipy stats use get the pearson corrlollation coefficient and use the text to put that on the graph.

Do the same for Band 7 vs Band 6. Which is more related?

```python
fig = plt.figure(figsize=(15,10))
gs = mpl.gridspec.GridSpec(1,1)
ax = fig.add_subplot(gs[0])

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

ax.xaxis.set_tick_params('major', length=5, labelsize=7)
ax.yaxis.set_tick_params('major', length=5, labelsize=7)

#plot data and customize look of plot
ax.plot(x_b1,y_b6, marker='.', color='m', linestyle='',
            markersize=3, clip_on=False, alpha=0.5)
#set axis limits
ax.set_xlim((min(x_b1),max(x_b1)))
ax.set_ylim((min(y_b6),max(y_b6)))

#scale the view to show all datapoints
ax.autoscale_view()

#set labels
ax.set_title("A comparison of Band 1 and Band 6 Reflectance", fontdict=None, loc='center')
ax.set_xlabel(r'Band 1 Reflectance', fontsize=10, family='sans-serif')
ax.set_ylabel('Band 6 Reflectance', fontsize=10, family='sans-serif')

coefficients = np.polyfit(x_b1, y_b6, 1)
plt.plot(x_b1, np.polyval(coefficients, x_b1), color='r', label='Linear Regression')
plt.legend()
slop=coefficients[0]
pearson = scipy.stats.pearsonr(x_b1, y_b6)
ax.text(0.4,0.6, "slope = " + str(slop), fontsize=12)
ax.text(0.4,0.57, "pearson coefficient= " + str(pearson), fontsize=12)

plt.show()
```

```python
b7 = data['B7']
x_b7 = b7.ravel()[samplesInds]
b6 = data['B6']
y_b6 = b6.ravel()[samplesInds]

fig = plt.figure(figsize=(15,10))
gs = mpl.gridspec.GridSpec(1,1)
ax = fig.add_subplot(gs[0])

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

ax.xaxis.set_tick_params('major', length=5, labelsize=7)
ax.yaxis.set_tick_params('major', length=5, labelsize=7)

#plot data and customize look of plot
ax.plot(x_b7,y_b6, marker='.', color='m', linestyle='',
            markersize=3, clip_on=False, alpha=0.5)
#set axis limits
ax.set_xlim((min(x_b7),max(x_b7)))
ax.set_ylim((min(y_b6),max(y_b6)))

#scale the view to show all datapoints
ax.autoscale_view()

#set labels
ax.set_title("A comparison of Band 7 and Band 6 Reflectance", fontdict=None, loc='center')
ax.set_xlabel(r'Band 7 Reflectance', fontsize=10, family='sans-serif')
ax.set_ylabel('Band 6 Reflectance', fontsize=10, family='sans-serif')

coefficients = np.polyfit(x_b7, y_b6, 1)
plt.plot(x_b7, np.polyval(coefficients, x_b7), color='r', label='Linear Regression')
plt.legend()
slop=coefficients[0]
pearson = scipy.stats.pearsonr(x_b7, y_b6)
ax.text(0.05,0.6, "slope = " + str(slop), fontsize=12)
ax.text(0.05,0.57, "pearson coefficient= " + str(pearson), fontsize=12)

plt.show()
```
